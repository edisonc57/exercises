```python
from math import *
import sympy as sp
import sympy.plotting as plot
```

```python
def Poltaylor(a,n):
  x=sp.symbols("x")
  f=sp.cos(x)
  #F= f
  T= f.subs(x,a)
  for k in range(1,n+1):
     dfk=sp.diff(f,x)
     T=T+dfk.subs(x,a)*((x-a)**k)/factorial(k)
     f=dfk

  print(sp.expand(T))
  g= plot(f,T,(x,a-3,a+3),title="Ряд Тейлора", show=False)
  g[0].line_color="k"
  g[1].line_color= "r"
  g.show()

a=float(input("Alrededor de: "))
n= int(input("Orden de "))
Poltaylor(a,n)
```
