```python
#interfacial tension сила межфазного натяжения
смн = 72
#angle of wettability контактный угол
ку= 45

#нам нужны другие типы единиц

смн= 72 * (1/68947.57)

k1= 1 * 9.869233e-12
k2= 15 * 9.869233e-12
k3= 100 * 9.869233e-12
k4= 500  * 9.869233e-12

пор1= 0.08
пор2= 0.11
пор3= 0.15
пор4= 0.22

```

```python
J1= (капиллярное.Pc_1/(смн*np.cos(np.deg2rad(ку))))*np.sqrt(k1/пор1)
J2= (капиллярное.Pc_2/(смн*np.cos(np.deg2rad(ку))))*np.sqrt(k2/пор2)
J3= (капиллярное.Pc_3/(смн*np.cos(np.deg2rad(ку))))*np.sqrt(k3/пор3)
J4= (капиллярное.Pc_4/(смн*np.cos(np.deg2rad(ку))))*np.sqrt(k4/пор4)

J= pd.DataFrame({"sw":капиллярное.sw, "Леверетт J1":J1,"Леверетт J2":J2, "Леверетт J3":J3, "Леверетт J4": J4})
J.head()
```

```python
J1_Функция= plt.plot(капиллярное.sw, J1)
J2_Функция= plt.plot(капиллярное.sw, J2)
J3_Функция= plt.plot(капиллярное.sw, J3)
J4_Функция= plt.plot(капиллярное.sw, J4)

plt.xlabel("Sw"), plt.ylabel ("Леверетт J Функция")
plt.title("Леверетт J Функция")
plt.legend(('1 md, porosity 0.08', '15 md, porosity 0.11', '100 md, porosity 0.15', '500 md, porosity 0.22'))
```
